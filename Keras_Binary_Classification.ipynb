```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("/content/data.csv")
```

```python
data
```

```python
data.drop([data.columns[0],data.columns[32]],axis=1,inplace=True)
```

```python
data
```

```python
data['diagnosis'].replace(to_replace=['B','M'],value=[0,1],inplace=True)
```

```python
data = np.array(data) 
```

```python
np.random.shuffle(data)
```

```python
data
```

```python
Y = data[:,0]
```

```python
Y
```

```python
X = data[:,1:]
```

```python
X
```

```python
X_dash = (X - np.mean(X,axis=0))/np.std(X,axis=0)
```

```python
X_dash
```

```python
from sklearn.model_selection import train_test_split
```

```python
X,X_test,Y,Y_test = train_test_split(X_dash,Y,test_size=0.10,random_state=42)
```

```python
X.shape
```

```python
X_test.shape
```

```python
Y.shape
```

```python
Y_test.shape
```

```python
X_train,X_cv,Y_train,Y_cv = train_test_split(X,Y,test_size=0.15,random_state=42)
```

```python
X_train.shape
```

```python
Y_train.shape
```

```python
X_cv.shape
```

```python
Y_cv.shape
```

```python
from keras.layers import InputLayer,Dense,BatchNormalization
from keras.models import Sequential
from keras.losses import binary_crossentropy
from keras.regularizers import l2
```

```python
def build_network():

  network = Sequential()
  network.add(InputLayer(input_shape=(X_train.shape[1],)))
  network.add(Dense(units=X_train.shape[1],activation="relu",kernel_initializer="glorot_normal",kernel_regularizer=l2(0.1)))
  network.add(BatchNormalization())
  network.add(Dense(units=1,activation="sigmoid",kernel_initializer="glorot_normal",kernel_regularizer=l2(0.1)))

  return network
```

```python
network = build_network()
```

```python
network.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])
```

```python
performance_history = network.fit(x=X_train,y=Y_train,batch_size=5,epochs=100,validation_data=(X_cv,Y_cv))
```

```python
network.summary()
```

```python
network.evaluate(x=X_test,y=Y_test)
```
